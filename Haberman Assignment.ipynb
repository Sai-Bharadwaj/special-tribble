```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Reading/ importing Habersman data

hbm =pd.read_csv('haberman.csv', names=['Age','Year of Treatment','Axilary Nodes', 'Survival Status'])
hbm
```

```python
#Objective is to find, 1- No of points 2- No of features 3-number of classes 4-data-points per class

#for knowing data points and features

print(hbm.shape)
print("Datapoints = 305", "4D Array")
print(hbm.columns)
```

```python
#For data points in the data set

print(hbm['Survival Status'].value_counts())
print(hbm['Year of Treatment'].value_counts())
print("\n")
print(hbm['Axilary Nodes'].value_counts())
```

```python
#2D Scatter plot
print(hbm.describe())
hbm.plot(kind = 'scatter', y = 'Age', x = 'Survival Status')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hbm, hue = 'Survival Status', size = 14) \
.map(plt.scatter, 'Survival Status', 'Age') \
.add_legend()

plt.show()

#Observations:

'''
1. There are 306 records, in which 225 people survived more that 5 years and 81 people survived within 5 years after treatment.
2. Avg age group is 52 years, and minimum age is 30 years and max age is 83.
3. more number of people survied for more than 5 years after treatment.
```

```python
#Histogram and PDF and CDF

'''
CDF

univariate analysis
'''

import numpy as np

sns.FacetGrid(hbm, hue="Survival Status", size=8) \
   .map(sns.distplot, "Age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(hbm, hue="Survival Status", size=8) \
   .map(sns.distplot, "Axilary Nodes") \
   .add_legend();
plt.show();

hbm.columns

sns.FacetGrid(hbm, hue="Survival Status", size=8) \
   .map(sns.distplot, "Year of Treatment") \
   .add_legend();
plt.show();

```

```python
#Building CDF


print("For Age: ")
counts, bin_edges = np.histogram(hbm['Age'], bins=10, density = True)
pdf = counts/(sum(counts))
print("PDF: ",pdf);
cdf = np.cumsum(pdf)
print("CDF: ",cdf)
print("Bind Edges: ",bin_edges);
plt.plot(bin_edges[1:], pdf, bin_edges[1:], cdf)
plt.xlabel("Age")
plt.title("CDF Vs PDF for AGE")
plt.show()


# For Year of treatment


print("For Year of Treatment: ")
counts, bin_edges = np.histogram(hbm['Year of Treatment'], bins=10, density = True)
pdf = counts/(sum(counts))
print("PDF: ",pdf);
cdf = np.cumsum(pdf)
print("CDF: ",cdf)
print("Bind Edges: ",bin_edges);
plt.plot(bin_edges[1:], pdf, bin_edges[1:], cdf)
plt.xlabel("Year of Treatment")
plt.title("CDF Vs PDF for Year of Treatment")
plt.show()

hbm.columns

#For Axilary Nodes

print("For Axilary Nodes: ")
counts, bin_edges = np.histogram(hbm['Axilary Nodes'], bins=10, density = True)
pdf = counts/(sum(counts))
print("PDF: ",pdf);
cdf = np.cumsum(pdf)
print("CDF: ",cdf)
print("Bind Edges: ",bin_edges);
plt.plot(bin_edges[1:], pdf, bin_edges[1:], cdf)
plt.xlabel("Axilary Nodes")
plt.title("CDF Vs PDF for Axilary Nodes")
plt.show()
```

```python
#Calculating Medians
print("Medians:-")
print("\n")
print("Median for Age:", np.median(hbm['Age']))
print("Median for Year of Treatment:", np.median(hbm['Year of Treatment']))
print("Median for Axilary Nodes:", np.median(hbm['Axilary Nodes']))
print("\n")

#Calculating Percentile
print("Percentiles:- ")
print("\n")
print("Percentile for Age:",np.percentile(hbm['Age'], np.arange(0,100,25)))
print("Percentile for Year of Treatment:",np.percentile(hbm['Year of Treatment'], np.arange(0,100,25)))
print("Percentile for Axilary Nodes:",np.percentile(hbm['Axilary Nodes'], np.arange(0,100,25)))
```
